```python
import requests, csv, re, pandas

APPSTORE_URL = "https://itunes.apple.com/search"

# Keywords
KEYWORDS = [
    "intimate partner violence"
    "abuse help"
    
]

COUNTRY = "us"   # we need to put some countries in it/ Plan A: language; Plan B: add countries manually; 
TOPK = 5         # 5 related apps (need discussion with team)

# keywordss（discussion needed）
IPV_HINTS = [
    r"intimate partner violence", r"domestic violence", r"IPV",
    r"abuse", r"victim", r"survivor", r"safety plan",
    r"hotline", r"helpline", r"violence prevention",
    r"gender[- ]based violence", r"relationship violence",
    r"counsel(ing)?", r"support( group)?"
]
IPV_RE = re.compile("|".join(IPV_HINTS), re.IGNORECASE)

def ios_search(keyword, country="us", limit=200):
    """调用 iTunes Search API，只取 App"""
    params = {"term": keyword, "entity": "software", "country": country, "limit": limit}
    r = requests.get(APPSTORE_URL, params=params, timeout=30)
    r.raise_for_status()
    data = r.json()
    return data.get("resultCount", 0), [x for x in data.get("results", []) if x.get("kind")=="software"]

def score_relevance(app, keyword):
    """打分：关键词/IPV词命中"""
    name = (app.get("trackName") or "").lower()
    desc = (app.get("description") or "").lower()
    kw = keyword.lower().strip()
    score = 0
    if kw in name: score += 5
    if kw in desc: score += 2
    if IPV_RE.search(name): score += 3
    if IPV_RE.search(desc): score += 1
    return score

def pick_top_related(results, keyword, topk=5):
    ranked = []
    for app in results:
        s = score_relevance(app, keyword)
        if s <= 0: continue
        ranked.append((
            s, app.get("userRatingCount") or 0, app.get("averageUserRating") or 0.0, app
        ))
    ranked.sort(key=lambda x: (x[0], x[1], x[2]), reverse=True)
    return [x[3] for x in ranked[:topk]]

# main 
all_rows = []
for kw in KEYWORDS:
    try:
        total, results = ios_search(kw, country=COUNTRY, limit=200)
        top_apps = pick_top_related(results, kw, topk=TOPK)
        if not top_apps:
            all_rows.append({
                "Keyword(s)": kw,
                "Operating system (iOS or Play Store)": "iOS",
                "Total number of search results": total,
                "Name of the IPV apps": "(No strong IPV-related hits)",
                "Rating": "",
                "Number of reviews": "",
                "Category set by the app Store": "",
                "Developer": "",
            })
        else:
            for i, app in enumerate(top_apps):
                all_rows.append({
                    "Keyword(s)": kw if i==0 else "",
                    "Operating system (iOS or Play Store)": "iOS" if i==0 else "",
                    "Total number of search results": total if i==0 else "",
                    "Name of the IPV apps": app.get("trackName"),
                    "Rating": app.get("averageUserRating"),
                    "Number of reviews": app.get("userRatingCount"),
                    "Category set by the app Store": app.get("primaryGenreName"),
                    "Developer": app.get("sellerName") or app.get("artistName"),
                })
    except Exception as e:
        all_rows.append({
            "Keyword(s)": kw,
            "Operating system (iOS or Play Store)": "iOS",
            "Total number of search results": "",
            "Name of the IPV apps": f"(Error: {e})",
            "Rating": "",
            "Number of reviews": "",
            "Category set by the app Store": "",
            "Developer": "",
        })

# print
outfile = "ipv_ios_results.csv"
with open(outfile, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=[
        "Keyword(s)",
        "Operating system (iOS or Play Store)",
        "Total number of search results",
        "Name of the IPV apps",
        "Rating",
        "Number of reviews",
        "Category set by the app Store",
        "Developer",
    ])
    writer.writeheader()
    for row in all_rows:
        writer.writerow(row)
        print(outfile)
```

```python
import requests, csv, re, time
from typing import List, Dict, Any, Tuple

APPSTORE_URL = "https://itunes.apple.com/search"

# Keywords (fixed comma)
KEYWORDS: List[str] = [
    "intimate partner violence",
    "abuse help",
]

# Search across multiple storefronts if you like
COUNTRIES: List[str] = ["us"]  # e.g., ["us", "ca", "gb", "au"]
TOPK = 5
LIMIT = 200  # Apple caps at 200

IPV_RE = re.compile("|".join(IPV_HINTS), re.IGNORECASE)

def ios_search(keyword: str, country: str = "us", limit: int = LIMIT) -> Tuple[int, List[Dict[str, Any]]]:
    """Call iTunes Search API; return (resultCount, list-of-apps)."""
    params = {"term": keyword, "entity": "software", "country": country, "limit": limit}
    r = requests.get(APPSTORE_URL, params=params, timeout=30)
    r.raise_for_status()
    data = r.json()
    results = [x for x in data.get("results", []) if x.get("kind") == "software"]
    return data.get("resultCount", len(results)), results

def score_relevance(app: Dict[str, Any], keyword: str) -> float:
    """Heuristic score: keyword/IPV hits, with stronger weight on title matches."""
    name = (app.get("trackName") or "")
    desc = (app.get("description") or "")
    kw = keyword.strip()

    # Case-insensitive lookups but preserve origin for regex
    name_l = name.lower()
    desc_l = desc.lower()
    kw_l = kw.lower()

    score = 0.0
    # Exact keyword signals
    if kw_l in name_l:
        score += 7.0
    if kw_l in desc_l:
        score += 3.0

    # IPV phrase hints
    if IPV_RE.search(name):
        score += 3.0
    if IPV_RE.search(desc):
        score += 1.0

    # Minor boost if placed in "Medical" or "Health & Fitness" or "Education"
    genre = (app.get("primaryGenreName") or "").lower()
    if genre in {"medical", "health & fitness", "education"}:
        score += 0.5

    return score

def pick_top_related(results: List[Dict[str, Any]], keyword: str, topk: int = TOPK) -> List[Dict[str, Any]]:
    ranked = []
    for app in results:
        s = score_relevance(app, keyword)
        if s <= 0:
            continue
        # Use stable, deterministic tie-breakers: rating count, avg rating, name
        user_cnt = app.get("userRatingCount") or app.get("userRatingCountForCurrentVersion") or 0
        avg = app.get("averageUserRating") or app.get("averageUserRatingForCurrentVersion") or 0.0
        ranked.append((s, int(user_cnt), float(avg), (app.get("trackName") or ""), app))
    ranked.sort(key=lambda x: (x[0], x[1], x[2], x[3].lower()), reverse=True)
    return [x[4] for x in ranked[:topk]]

# Main
all_rows: List[Dict[str, Any]] = []
seen_ids = set()  # to dedupe across keywords/countries

for country in COUNTRIES:
    for kw in KEYWORDS:
        try:
            total, results = ios_search(kw, country=country, limit=LIMIT)
            top_apps = pick_top_related(results, kw, topk=TOPK)

            # If none pass the threshold, still record the search meta
            if not top_apps:
                all_rows.append({
                    "Keyword(s)": kw,
                    "Country": country.upper(),
                    "Operating system (iOS or Play Store)": "iOS",
                    "Total number of search results": total,
                    "Name of the IPV apps": "(No strong IPV-related hits)",
                    "Rating": "",
                    "Number of reviews": "",
                    "Category set by the App Store": "",
                    "Developer": "",
                    "App Store URL": "",
                    "Bundle ID": "",
                })
            else:
                first = True
                for app in top_apps:
                    tid = app.get("trackId")
                    if tid in seen_ids:
                        continue
                    seen_ids.add(tid)

                    row_prefix = {
                        "Keyword(s)": kw if first else "",
                        "Country": country.upper() if first else "",
                        "Operating system (iOS or Play Store)": "iOS" if first else "",
                        "Total number of search results": total if first else "",
                    }
                    first = False

                    all_rows.append({
                        **row_prefix,
                        "Name of the IPV apps": app.get("trackName") or "",
                        "Rating": app.get("averageUserRating")
                                  or app.get("averageUserRatingForCurrentVersion") or "",
                        "Number of reviews": app.get("userRatingCount")
                                             or app.get("userRatingCountForCurrentVersion") or "",
                        "Category set by the App Store": app.get("primaryGenreName") or "",
                        "Developer": app.get("sellerName") or app.get("artistName") or "",
                        "App Store URL": app.get("trackViewUrl") or "",
                        "Bundle ID": app.get("bundleId") or "",
                    })

            # small pause to be courteous if you expand country list
            time.sleep(0.2)

        except Exception as e:
            all_rows.append({
                "Keyword(s)": kw,
                "Country": country.upper(),
                "Operating system (iOS or Play Store)": "iOS",
                "Total number of search results": "",
                "Name of the IPV apps": f"(Error: {e})",
                "Rating": "",
                "Number of reviews": "",
                "Category set by the App Store": "",
                "Developer": "",
                "App Store URL": "",
                "Bundle ID": "",
            })

# Write CSV
# instead of writing to CSV:
for row in all_rows:
    print("Keyword(s):", row["Keyword(s)"])
    if "Country" in row:
        print("Country:", row["Country"])
    print("Operating system:", row["Operating system (iOS or Play Store)"])
    print("Total results:", row["Total number of search results"])
    print("App name:", row["Name of the IPV apps"])
    print("Rating:", row["Rating"])
    print("Reviews:", row["Number of reviews"])
    print("Category:", row["Category set by the App Store"])
    print("Developer:", row["Developer"])
    if "App Store URL" in row:
        print("URL:", row["App Store URL"])
    print("-" * 60)

```
